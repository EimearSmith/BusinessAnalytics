```python
#Import a CSV file as a Pandas Dataframe.
import pandas as pd
House_Prices=pd.read_csv("Property_Price_Register_Ireland-28-05-2021.csv")
print(House_Prices.head())
```

```python
# Pull in entire data from CSV which shows there are 476,745 rows and 9 columns
House_Prices
```

```python
#here we can explore the DataFrame and see what address has the most number of sales
House_Prices.groupby(['ADDRESS']).count()['COUNTY'].sort_values(ascending=False)
```

```python
#Another groupby on different colums shows us that despite the more houses sold in Midleton, the highest average price was in Dublin
House_Prices.groupby(['COUNTY']).mean()['SALE_PRICE'].sort_values(ascending=False)
```
